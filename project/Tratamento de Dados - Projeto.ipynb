```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
import time
import statistics
```

# Tratamento dos Dados na Head

```python
country=[]
continent=[]
population=[]
weekly_count=[]
cumulative_count=[]
year_week=[]
indicator=[]

with open('banco_de_dados.csv', newline='') as dist:
  leitor = csv.DictReader(dist)
  for key in leitor:
    for i in range(len(key)):
      country.append(key['\ufeffcountry'])
      continent.append(key['continent'])
      indicator.append(key['indicator'])
      population.append(int(key['population']))
      weekly_count.append(int(key['weekly_count']))
      cumulative_count.append(int(key['cumulative_count']))
      year_week.append(key['year_week'])
```

# Tratamento de Dados para os Casos

Serão tratados os casos de COVID-19 no continente asiático no código abaixo:

```python
#Tratamento com os países asiáticos (Casos)

Asia_value=[]
Asia_key=[]
Asia_country=[]
pais_inicial = country[0]
aux_value=[]
aux_key=[]
aux_country=[]
j=0

for i in range(len(country)):
  if country[i] == pais_inicial and continent[i]=='Asia' and indicator[i] == 'cases':
    aux_key.append(j)
    aux_value.append(weekly_count[i])
    aux_country.append(country[i])
    j=j+1
  if country[i] != pais_inicial and continent[i]=='Asia' and indicator[i] == 'cases':
    j=0
    Asia_key.append(aux_key)
    Asia_value.append(aux_value)
    Asia_country.append(aux_country)
    aux_value=[]
    aux_key=[]
    aux_country=[]
    aux_key.append(j)
    aux_value.append(weekly_count[i])
    aux_country.append(country[i])
    j=j+1
    pais_inicial = country[i]
```

Será plotado no código abaixo a pasta com todas as imagens dos gráficos.

```python
#Plot com cada um dos países asiáticos (Casos)

pais_inicial = Asia_country[0][0]
lista_x = []
lista_y = []
nome = ''

for i in range(len(Asia_key)):
  nome=Asia_country[i][0]
  lista_x = Asia_key[i]
  lista_y = Asia_value[i]
  fig = plt.figure()
  plt.xticks(rotation=90)
  plt.plot(lista_x, lista_y,'o')
  plt.savefig('Imagens_Casos/'+nome+'(Casos).png', bbox_inches = "tight")
  plt.show()
```

```python
def gauss(x,mu,sigma,A):
    return A*exp(-(x-mu)**2/2/sigma**2)
```

```python
func_normal1 = []
func_normal2 = []
funcao=0
tam = len(Asia_key[1])
lista_x1 = []
lista_y1 = []
lista_x2 = []
lista_y2 = []
f1=[]

for i in range(0, 620):
  lista_x1.append(Asia_key[1][i])
  lista_y1.append(Asia_value[1][i])

for i in range(620, tam):
  lista_x2.append(Asia_key[1][i])
  lista_y2.append(Asia_value[1][i])

lista_x = lista_x1 + lista_x2
lista_y = lista_y1 + lista_y2

media = statistics.mean(lista_y)
variancia = statistics.pvariance(lista_y)

media1 = statistics.mean(lista_y1)
variancia1 = statistics.pvariance(lista_y1)

media2 = statistics.mean(lista_y2)
variancia2 = statistics.pvariance(lista_y2)

A = 1/(variancia1*np.sqrt(2*np.pi))

for i in range(0, 620):
  f1.append(gauss(lista_x1[i], media1, variancia1, A))

f2=[]
A = 1/(variancia2*np.sqrt(2*np.pi))

for i in range(0, tam-620):
  f2.append(gauss(lista_x2[i], media2, variancia2, A))


plt.plot(f1, lista_y1)
plt.ylabel('gaussian distribution')
plt.savefig('normal1.png', bbox_inches = "tight")
plt.show()

plt.plot(f2, lista_y2)
plt.ylabel('gaussian distribution')
plt.savefig('normal2.png', bbox_inches = "tight")
plt.show()
```
